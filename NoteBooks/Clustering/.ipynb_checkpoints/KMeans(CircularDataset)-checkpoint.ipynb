# K-Means - Circular Data

```python
# Importing the dependencies
import math

import pandas as pd
import numpy as np

from sklearn.cluster import KMeans

import matplotlib.pyplot as plt
import plotly.express as px
```

### Creating the dataset


```python
def circle(r ,n , noise=20, seed=42):
    np.random.seed(seed)
    points = []
    for x in range(1, n+1):
        point = (math.cos(2 * math.pi / n * x) * r + np.random.normal(-noise, noise),
                math.sin(2 * math.pi / n * x) * r + np.random.normal(-noise, noise))
        
        points.append(point)
        
    return points

def list_to_df(ls, i):
    df = pd.DataFrame(ls)
    df["cluster"] = str(i + 1)
    return df


small_circle = circle(100, 300, 10)
medium_circle = circle(300, 700, 20)
big_circle = circle(500, 1000, 30)

noise = [(np.random.randint(-600, 600),
         np.random.randint(-600, 600)) for i in range(300)]


data = [list_to_df(ls, i) for i, ls in enumerate([small_circle, medium_circle, big_circle, noise])]
data = pd.concat(data)
data.columns = ['x', 'y', 'cluster']
data.sample(10)
```

```python
#### In case you want to save the data as csv

data.to_csv("data/circular_data.csv", index=False)
```

### Visualizing the data

```python
# Scatter plot with plotly
fig = px.scatter(data,
                    x="x",
                    y="y",
                    color="cluster",
                    height=500)
fig.show()
```

```python
#  Selecting the features for training
X = data[["x","y"]]
```

### KMeans model training

```python
# Instantiating the model

kmeans = KMeans(n_clusters=3, random_state=0, n_init="auto")

# Fitting the model with the selected features
kmeans.fit(X)
```

```python
# Predicting with the fitted model
data["predicted_cluster"] = kmeans.predict(X)

# Showing a sample with the predicted clusters
data.sample(6)
```

## Results Visualization


```python
fig = px.scatter(data,
                    x="x",
                    y="y",
                    color="predicted_cluster",
                    height=500)
fig.show()
```

### Analysis of the model behavior

Using standard K-means clustering might not be appropriate because it doesn't account for the circular nature of the data. In circular data, the distance between values is not linear.
