# K-Means - Iris Dataset

```python
import pandas as pd
import numpy as np

from sklearn.datasets import load_iris

from sklearn.cluster import KMeans
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, f1_score

import matplotlib.pyplot as plt
import plotly.express as px
```

```python
# General config of pandas
pd.set_option("display.max_rows", None)
pd.set_option("display.max_columns", None)
pd.set_option("display.width", None)
pd.set_option('display.expand_frame_repr', False)
```

```python
# Creating the dataframe with iris dataset

data = load_iris()

df = pd.DataFrame(data = data.data, columns=data.feature_names)

df.head()
```

```python
# Adding the target column so that we can do a comparison of the results 
df["class_code"] = data.target

df["class_name"] = data.target_names[df["class_code"].astype(int)]

df.sample(10)
```
